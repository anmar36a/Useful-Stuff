```python
img_url_sample = "https://onepiece-fans2.net/mangafiles/es/full-color/836/147.jpg"
```

```python
def get_image_size(url):
    """Gets the dimensions of a JPEG image from a URL without fully downloading it."""
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()

        image = Image.open(BytesIO(response.content))
        # Only load the header to get dimensions without loading the full image
        image.load()
        return image.width, image.height

    except (requests.exceptions.RequestException, IOError) as e:
        print(e)
        return None, None
```

```python
get_image_size(img_url_sample)
```

```python
def count_pages(chapter_url):
    """Count the number of pages, accounting for double-page spreads."""
    page_count = 0
    page_left = True
    page = 1
    while page_left:
        page_str = f"0{page}" if len(str(page)) == 1 else str(page)
        img_url = f"{chapter_url}/{page_str}.jpg"

        try:
            width, height = get_image_size(img_url)
            if width > 300:
              if width > 1.8 * height:  # Heuristic for a double-page spread
                  page_count += 2
              else:
                  page_count += 1
            else: page_left = False
        except Exception as e:
            print(f"Error processing image {img_url}: {e}")
    print("eureka")
    return page_count
```

```python
count_pages("https://onepiece-fans2.net/manga/es/full-color/96")
```

```python
def main():
    base_url = "https://onepiece-fans2.net/manga/es/full-color/"
    output_file = "one_piece_chapters.csv"

    # Open the CSV file for writing
    with open(output_file, mode='w', newline='', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(["Chapter", "Pages"])

        # Iterate through chapters 1 to 1000
        for chapter in range(1, 1001):
            print(f"Processing Chapter {chapter}...")
            if chapter < 10:
              chapter_url = f"{base_url}{chapter:02d}"
            else:
              chapter_url = f"{base_url}{chapter}"  # Ensure 4 digits, e.g., 0001
            try:
                page_count = count_pages(chapter_url)
                writer.writerow([chapter, page_count])
                print(f"Chapter {chapter} has {page_count} pages.")
            except Exception as e:
                print(f"Error processing Chapter {chapter}: {e}")

    print(f"Data saved to {output_file}")
```

```python
if __name__ == "__main__":
    main()
```
